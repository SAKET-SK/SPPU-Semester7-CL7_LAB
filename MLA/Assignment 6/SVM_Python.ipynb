#### Importing the libraries and datasets

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

# Import dataset from Sklearn library (Breast cancer dataset bu UCI)

from sklearn.datasets import load_breast_cancer
cancer_dataset = load_breast_cancer()
```

## Exploring data

```python
print("Features:", cancer_dataset.feature_names)
print("Labels: ", cancer_dataset.target_names)
```

```python
cancer_df = pd.DataFrame(np.c_[cancer_dataset['data'], cancer_dataset['target']], 
                         columns = np.append(cancer_dataset['feature_names'], ['target']))
cancer_df.head()
```

```python
# Information of the dataset
cancer_df.info()
```

```python
# Dimensions of the dataframe
cancer_df.shape
```

```python
# Describe the dataset
cancer_df.describe()
```

```python
# Check if there are null values
cancer_df.isnull().sum()
```

```python
# ip (input) will contain all the columns other than 'target'.
ip = cancer_df.drop(['target'], axis = 1)
ip.head()
```

```python
# op (output) will contain only the target values
op = cancer_df['target']
op.head()
```

```python
# Importing train_test_split for splitting the data into training and testing part
from sklearn.model_selection import train_test_split
```

#### Dividing dataset into 70% for training and 30% for testing

```python
ip_train, ip_test, op_train, op_test = train_test_split(ip, op, test_size = 0.3, random_state = 20)
```

```python
# Checking the shape of the training and testing data
print("The shape of training input is " + str(ip_train.shape))
print("The shape of testing input is " + str(ip_test.shape))
print("The shape of training output is " + str(op_train.shape))
print("The shape of testing output is " + str(op_test.shape))
```

### Import SVM model

```python
from sklearn.svm import SVC
svc_model = SVC()
```

```python
svc_model.fit(ip_train, op_train)
```

### Predict using the trained model

```python
op_predict = svc_model.predict(ip_test)
```

### Creating confusion matrix for checking the output

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
mat = np.array(confusion_matrix(op_test, op_predict, labels=[1, 0]))
confusion = pd.DataFrame(mat, index=['is cancer', 'is healthy'],
                        columns=['predicted cancer', 'predicted_healthy'])
confusion
```

```python
print(classification_report(op_test, op_predict))
```

```python
sns.heatmap(confusion, annot=True, cmap="YlGnBu")
```

```python
from sklearn import metrics
```

```python
print("Accuracy of the model is: ",metrics.accuracy_score(op_test, op_predict))
```
